# Iris Classification

The Dataset contains 3 classes of 50 instances each, Where each class refers to type of iris plant. one class linearly separable from the other 2.
The latter are not linearly separable from each other.

Attribute Information

1. Sepal length in cm
2. Sepal  width in cm
3. Petal length in cm
4. Petal  width in cm
5. Class:- Iris Setosa -- Iris Versicolor -- Iris Verginica

# Imported modules

```python
import warnings

warnings.filterwarnings('ignore')

import pandas as pd

import numpy as np
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Species"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(df,test_size=0.25)
```

```python
train
```

```python
test
```

```python
train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
train_Y = train[['Species']]
```

```python
train_X
```

```python
train_Y
```

```python
test_X=test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
test_Y=test[['Species']]
```

```python
test_X
```

```python
test_Y
```

# Support Vector Machine Use For Accuracy.

```python
from sklearn import svm
```

```python
#svm
model=svm.SVC()
```

```python
model.fit(train_X,train_Y)
pred = model.predict(test_X)
```

```python
pred
```

```python
from sklearn  import metrics
```

```python
metrics.accuracy_score(pred,test_Y)*100
```

# Decision Tree to find accuracy.

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(train_X,train_Y)
pred = model.predict(test_X)
```

```python
metrics.accuracy_score(pred,test_Y)*100
```

# KNearestNeighbor to find accuracy

```python
#KNN 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(train_X,train_Y)
pred=knn.predict(test_X)
```

```python
metrics.accuracy_score(pred,test_Y)*100
```
